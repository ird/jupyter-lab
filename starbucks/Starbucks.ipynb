```python
import pandas as pd

""" prepare the data """
stores = pd.read_csv('directory.csv')
lon = stores[(stores.Country == 'GB') & (stores.City == 'London') & (stores.Latitude < 51.6)]
points = [[x, y] for x,y in lon.iloc[:, 11:].values] # extract lat and long
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
v = Voronoi(points)
voronoi_plot_2d(v)
plt.rcParams['figure.figsize'] = (9.2, 7.8)
plt.show()
```
